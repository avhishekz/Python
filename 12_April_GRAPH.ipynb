```python
plt.rcParams["figure.figsize"] = (5,3)   #fig dimensions
```

```python
plt.plot(batsman["index"], batsman["V Kohli"])
plt.plot(batsman["index"], batsman["RG Sharma"])
plt.title("Rohit Sharma vs Virat Kohli Comparison of IPL Runs")
plt.xlabel("season")
plt.ylabel("Runs scored")
```

```python
plt.plot(batsman["index"], batsman["V Kohli"],color="#9BEE12" , linestyle="dotted", linewidth=3)
plt.plot(batsman["index"], batsman["RG Sharma"],color="blue" , linestyle="dotted", linewidth=2)
plt.title("Rohit Sharma vs Virat Kohli Comparison of IPL Runs")
plt.xlabel("season")
plt.ylabel("Runs scored")
```

```python
plt.plot(batsman["index"], batsman["V Kohli"],color="#9BEE12" , linestyle="dotted", linewidth=3 ,marker="D")
plt.plot(batsman["index"], batsman["RG Sharma"], color="#0000FF", linestyle="dotted", linewidth=2, marker="D")

plt.xlabel("season")
plt.ylabel("Runs scored")
```

```python
plt.plot(batsman["index"], batsman["V Kohli"], color="#9BEE12", linestyle="dotted", linewidth=3, marker="D", label="Virat")
plt.plot(batsman["index"], batsman["RG Sharma"], color="#0000FF", linestyle="dotted", linewidth=2, marker="D", label="Rohit")
plt.title("Rohit Sharma vs Virat Kohli Comparison of IPL Runs")
plt.xlabel("Season")
plt.ylabel("Runs scored")
plt.legend(loc="upper right")
```
