```python
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)
n = int(input("Enter a non-negative integer: "))
print(f"The factorial of {n} is {factorial(n)}")
```

```python
def GCD(x,y)
  rem = x%y
  if rem==0:
   return y
  else:
    return GCD(y,rem)
```

```python
def fib(n):
  if n<2:
    return n
  else:
    return fib (n-1)+fib(n-2)

n=int(input("Enter the number of terms"))
for i in range(n):
  print(fib(i))
```
